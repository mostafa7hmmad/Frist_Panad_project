# What city had highest number of sales

### Frist WE need To Get City From Adress this Way hepl to do t

```python
def extract_city(address):
    # Split the address by commas
    parts = address.split(', ')
    if len(parts) > 1:
        return parts[1]
    return None

# Apply the function to create a new 'City' column
df['City'] = df['Purchase Address'].apply(extract_city)
df.to_csv('C:/Users/mostafa/AI&ML/CSV/all_data_copy.csv',index=False)
```

### the City had highest sales

```python
Data=df.groupby('City').sum()
Data
```

```python
plt.figure(figsize=(10, 6))
City=df['City'].unique()
plt.bar(City, Data['Sales'],color='#228B22')
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
plt.yticks(Data['Sales'])  # Optional: Rotate x-axis labels for better readability
plt.ylabel('This is Sales ($)')
plt.xlabel('This is City')
plt.title('Sales by City')
plt.tight_layout() 
```
